```python
""" 
module to convert saved models to tflite format. requires tf-nightly==2.4.0-dev20200803

- https://github.com/tensorflow/models/blob/master/research/object_detection/g3doc/running_on_mobile_tf2.md
- https://github.com/tensorflow/models/blob/master/research/object_detection/g3doc/tf2_detection_zoo.md
"""
from typing import List
import os

import tensorflow as tf

print(tf.__version__)

def convert_saved_models(models_dir: str, saved_models: List[str]) -> None:
    for saved_model in saved_models:
        print(f"converting model: {saved_model}")
        # https://github.com/tensorflow/models_dir/issues/41877
        path_to_saved_model = os.path.join(MODELS_DIR, saved_model, "tflite", "saved_model")
#         path_to_saved_model = os.path.join(MODELS_DIR, saved_model, "saved_model")
        print("path_to_saved_model: ", path_to_saved_model)

        converter = tf.lite.TFLiteConverter.from_saved_model(path_to_saved_model, signature_keys=['serving_default'])

        converter.optimizations = [tf.lite.Optimize.DEFAULT]
        converter.experimental_new_converter = True
        converter.target_spec.supported_ops = [
            tf.lite.OpsSet.TFLITE_BUILTINS,
#             tf.lite.OpsSet.TFLITE_BUILTINS_INT8,
            tf.lite.OpsSet.SELECT_TF_OPS
        ]

        tflite_model = converter.convert()

        # Save the TF Lite model.
        filename = os.path.join(path_to_saved_model, "detect.tflite")
        print(f'saving model: {filename}')
        with tf.io.gfile.GFile(filename, 'wb') as f:
            f.write(tflite_model)
        print(f'saved model: {filename}')

    print("done ...")
```

```python
MODELS_DIR = "../models"
# MODELS_DIR = "/tmp"
SAVED_MODELS = [
    ### https://github.com/tensorflow/models/blob/master/research/object_detection/g3doc/tf2_detection_zoo.md ###
#     "ssd_resnet50_v1_fpn_640x640_coco17_tpu-8",
#     "ssd_mobilenet_v2_fpnlite_640x640_coco17_tpu-8",
    "ssd_mobilenet_v2_fpnlite_320x320_coco17_tpu-8",
    "ssd_mobilenet_v2_320x320_coco17_tpu-8",
#     "ssd_mobilenet_v2_quantized_300x300_coco_2019_01_03"
    ### https://github.com/tensorflow/models/blob/master/research/object_detection/g3doc/tf1_detection_zoo.md ###
#     "ssd_mobilenet_v1_fpn_shared_box_predictor_640x640_coco14_sync_2018_07_03",

]
```

```python
convert_saved_models(MODELS_DIR, SAVED_MODELS)

# run from project root
# !cd .. && ./bin/tensorflow/run_tflite_convert.sh ssd_mobilenet_v2_fpnlite_320x320_coco17_tpu-8

print('DONE!')

```
